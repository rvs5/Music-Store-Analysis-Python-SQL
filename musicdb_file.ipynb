2.Which country have the most invoices

```python
%%sql

Select billing_country, count(*) as no_of_invoice from invoice

group by billing_country
order by count(*) desc
;
```

3.What are the top 3 value of total invoice

```python
%%sql
select total from invoice

order by total desc
limit 3;
```

4.Which city has best customers

```python
%%sql
select billing_city ,sum(total) as Total from invoice
group by billing_city
order by Total desc
;
```

5.Who is the Best customer

```python
%%sql 

select customer.customer_id,customer.first_name,customer.last_name , sum(invoice.total) as Total
from customer join Invoice 
on
customer.customer_id = Invoice.customer_id
group  by customer.customer_id
order by Total desc
limit 1
;
```

6.Return email,firstname, lastname & genre of all Rock music listners return your list ordered aplhabetically by email starting with A

```python
%%sql

select distinct email,first_name,last_name
from customer 
join invoice on customer.customer_id = invoice.customer_id
join invoice_line on invoice.invoice_id = invoice_line.invoice_id
where track_id in (

    select track_id from track
    join genre on track.genre_id = genre.genre_id
    where genre.name like 'Rock'
)

order by email asc
;
```

7. Now, Return the artist who have written th most rock music in our dataset.

```python
%%sql

select artist.artist_id,artist.name,count(artist.artist_id) as no_of_songs from track
join album on track.album_id = album.artist_id 
join artist on album.artist_id = artist.artist_id
join genre on genre.genre_id = track.genre_id
where genre.name like 'Rock'
group by artist.artist_id
order by no_of_songs desc
limit 10

;
```

8.Return all the track names that have a song length longer than the average song length.

```python
# Here the avg is 393599.212103910933 to find avg -> (select avg(milliseconds) from track)

%%sql

select name, milliseconds from track 
where milliseconds > 393599.212103910933
order by milliseconds desc
limit 5;
```

9. number of purchases made by each country and their top famous genre name?

```python
%%sql

With popular_genre as (

select Count(customer.customer_id) as purchases,
customer.country, genre.name,
row_number() over(partition by customer.country
order by count(customer.customer_id) desc ) as row_no

from invoice_line
join invoice on invoice_line.invoice_id = invoice.invoice_id
join customer on invoice.customer_id = customer.customer_id
join track on invoice_line.track_id = track.track_id
join genre on track.genre_id = genre.genre_id

group by 2,3
order by 1 desc
)

select * from popular_genre where row_no <=1;
```
